# Hypothesis One: Gender might affect salary levels or positions within the company.

```python

#Python query to count the number of male and female employees
df_7 = pd.read_csv('second_query_2.csv')
df_7
```

```python
#Calculate average salary by gender
df_3 = pd.read_csv('third_query.csv')
df_3
```

```python
#barplot for the average salary by gender
plt.figure(figsize=(8, 4))
sns.barplot(x='gender', y='average_salary', data=df_3, palette='pastel')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```
