# Merge both dataframes on the ID column

```python

merged_df= pd.merge(employee_df, hr_df, on='ID', how='inner')
merged_df.head(50)
```

```python
#change the order of the column
desired_order= ["ID","Employee", "Gender", "Birth Date", "Education", "City", 
         "Position", "Performance Review", "Salary",
         "Hire Date", "Last Promotion Date", "Termination Date", "Termination Reason", 
         "Overdue Vacation?", "Department"]
df = merged_df.reindex(columns=desired_order)
```

```python
#drop columns unimportant for the analysis
columns_to_drop =["Last Promotion Date", "Termination Date", "Termination Reason"]
dropped_df = df.drop(columns=columns_to_drop)
```

```python
# Convert column names: lowercase and replace spaces with underscores
dropped_df.columns = [col.lower().replace(' ', '_') for col in dropped_df.columns]
```
